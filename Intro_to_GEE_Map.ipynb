## working with feature collection

```python
#set up a new map object
Map = geemap.Map()
Map
```

```python
# import uttarakhand shapefile layer
UK = ee.FeatureCollection('users/geographymanas/Uttarakhand')
UK_style = {'color': 'black',
            'fillColor': '00000000',
            'width': 1}
Map.addLayer(UK, UK_style, 'Uttarakhand')
Map.addLayer(UK, {}, 'UK.style (**UK_style)')
Map.centerObject(UK,7)
```

```python
#spatial filtering
#read gpw data from NASA SEDAC
gpw = ee.FeatureCollection('projects/sat-io/open-datasets/sedac/gpw-v4-admin-unit-center-points-population-estimates-rev11')

#filter the feature collection with only UK
#filter the points where population is more that 500000 in 2020
gpw_uk = gpw.filterBounds(UK)
Map.addLayer(gpw_uk, {}, 'gpw_uttarakhand')

gpw_uk_gt500k = gpw_uk.filter(ee.Filter.gt('UN_2020_E', 500000))
gpw_uk_gt500k_style = {'color': 'blue',
                       'fillColor': '00000000',
                       'width': 1 ,
                       'pointSize': 5}

Map.addLayer(gpw_uk_gt500k.style(**gpw_uk_gt500k_style), {}, 'gpw_uttarakhand_gt500k')
```

```python
#download the filtered feature collection
task = ee.batch.Export.table.toDrive(collection= gpw_uk_gt500k,
                                     description= 'gpw_uttarakhand_2020_500k',
                                     folder= 'GEE',
                                     fileFormat= 'SHP',
                                     fileNamePrefix= 'gpw_uttarakhand_2020_500k')
#export the filtered feature collection
#task.start()    #after runninf start , need to comment out it other its will run again
```
